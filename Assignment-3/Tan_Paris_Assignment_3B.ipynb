````python
# cleaning phonenumbers

# df.loc[:,"phonenumber"].replace(r"^0","+63",regex=True, inplace=True) # unnecessary
df.loc[:,"phonenumber"].replace(r"^917","+63 917",regex=True, inplace=True)
df.loc[:,"phonenumber"].replace(r"^918","+63 918",regex=True, inplace=True)
df.loc[:,"phonenumber"].replace(r"^927","+63 927",regex=True, inplace=True)
````

````python
df
````

````python
# birthmonths manual

# birthmonths in numbers

df.loc[:,"birthmonth"].replace(r"^2","February",regex=True, inplace=True)
df.loc[:,"birthmonth"].replace(r"^3","March",regex=True, inplace=True)
df.loc[:,"birthmonth"].replace(r"^4","April",regex=True, inplace=True)
df.loc[:,"birthmonth"].replace(r"^5","May",regex=True, inplace=True)
df.loc[:,"birthmonth"].replace(r"^6","June",regex=True, inplace=True)
df.loc[:,"birthmonth"].replace(r"^7","July",regex=True, inplace=True)
df.loc[:,"birthmonth"].replace(r"^8","August",regex=True, inplace=True)
df.loc[:,"birthmonth"].replace(r"^9","September",regex=True, inplace=True)
df.loc[:,"birthmonth"].replace(r"^10","October",regex=True, inplace=True)
df.loc[:,"birthmonth"].replace(r"^11","November",regex=True, inplace=True)
df.loc[:,"birthmonth"].replace(r"^12","December",regex=True, inplace=True)
df.loc[:,"birthmonth"].replace(r"^1","January",regex=True, inplace=True)

# birthmonths (full)

df.loc[:,"birthmonth"].replace(r"February","February",regex=True, inplace=True)
df.loc[:,"birthmonth"].replace(r"March","March",regex=True, inplace=True)
df.loc[:,"birthmonth"].replace(r"April","April",regex=True, inplace=True)
df.loc[:,"birthmonth"].replace(r"May","May",regex=True, inplace=True)
df.loc[:,"birthmonth"].replace(r"June","June",regex=True, inplace=True)
df.loc[:,"birthmonth"].replace(r"July","July",regex=True, inplace=True)
df.loc[:,"birthmonth"].replace(r"August","August",regex=True, inplace=True)
df.loc[:,"birthmonth"].replace(r"September","September",regex=True, inplace=True)
df.loc[:,"birthmonth"].replace(r"October","October",regex=True, inplace=True)
df.loc[:,"birthmonth"].replace(r"November","November",regex=True, inplace=True)
df.loc[:,"birthmonth"].replace(r"December","December",regex=True, inplace=True)
df.loc[:,"birthmonth"].replace(r"January","January",regex=True, inplace=True)

# birthmonths (short)

df.loc[:,"birthmonth"].replace(r"Feb$","February",regex=True, inplace=True)
df.loc[:,"birthmonth"].replace(r"Mar$","March",regex=True, inplace=True)
df.loc[:,"birthmonth"].replace(r"Apr$","April",regex=True, inplace=True)
df.loc[:,"birthmonth"].replace(r"Jun$","June",regex=True, inplace=True)
df.loc[:,"birthmonth"].replace(r"Jul$","July",regex=True, inplace=True)
df.loc[:,"birthmonth"].replace(r"Aug$","August",regex=True, inplace=True)
df.loc[:,"birthmonth"].replace(r"Sep$","September",regex=True, inplace=True)
df.loc[:,"birthmonth"].replace(r"Oct$","October",regex=True, inplace=True)
df.loc[:,"birthmonth"].replace(r"Nov$","November",regex=True, inplace=True)
df.loc[:,"birthmonth"].replace(r"Dec$","December",regex=True, inplace=True)
df.loc[:,"birthmonth"].replace(r"Jan$","January",regex=True, inplace=True)
````

````python
df
````

````python
# courses (manual)

df.loc[:,"course"].replace(r"BS Legal Management$","BS Legal Management",regex=True, inplace=True)
df.loc[:,"course"].replace(r"Legal Management$","BS Legal Management",regex=True, inplace=True)
df.loc[:,"course"].replace(r"BS LM$","BS Legal Management",regex=True, inplace=True)
df.loc[:,"course"].replace(r"LM$","BS Legal Management",regex=True, inplace=True)

df.loc[:,"course"].replace(r"BS Communications Technology Management$","BS Communications Technology Management",regex=True, inplace=True)
df.loc[:,"course"].replace(r"Communications Technology Management$","BS Communications Technology Management",regex=True, inplace=True)
df.loc[:,"course"].replace(r"Communications Technology$","BS Communications Technology Management",regex=True, inplace=True)
df.loc[:,"course"].replace(r"BS CTM$","BS Communications Technology Management",regex=True, inplace=True)
df.loc[:,"course"].replace(r"CTM$","BS Communications Technology Management",regex=True, inplace=True)

df.loc[:,"course"].replace(r"BS Management Engineering$","BS Management Engineering",regex=True, inplace=True)
df.loc[:,"course"].replace(r"Management Engineering$","BS Management Engineering",regex=True, inplace=True)
df.loc[:,"course"].replace(r"BS ME$","BS Management Engineering",regex=True, inplace=True)
df.loc[:,"course"].replace(r"ME$","BS Management Engineering",regex=True, inplace=True)

df.loc[:,"course"].replace(r"BS Information Technology Entrepreneurship$","BS Information Technology Entrepreneurship",regex=True, inplace=True)
df.loc[:,"course"].replace(r"Information Technology Entrepreneurship$","BS Information Technology Entrepreneurship",regex=True, inplace=True)
df.loc[:,"course"].replace(r"BS ITE$","BS Information Technology Entrepreneurship",regex=True, inplace=True)
df.loc[:,"course"].replace(r"ITE$","BS Information Technology Entrepreneurship",regex=True, inplace=True)

df.loc[:,"course"].replace(r"BS Management$","BS Management",regex=True, inplace=True)
df.loc[:,"course"].replace(r"^Management$","BS Management",regex=True, inplace=True)
df.loc[:,"course"].replace(r"BS MGT$","BS Management",regex=True, inplace=True)
df.loc[:,"course"].replace(r"MGT$","BS Management",regex=True, inplace=True)

````
