````python
dirty_birthyear_list = list()

with open('recweekdata.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            pass
        else:
            dirty_birthyear_list.append(row[4])
        line_count += 1
        
# dirty_birthyear_list

unique_birthyears = set(dirty_birthyear_list)

# unique_birthyears

[n for n in unique_birthyears]

unique_birthyears_dictionary = {name:None for name in unique_birthyears}

# unique_birthyears_dictionary

for i in unique_birthyears_dictionary:
    final_name = input(i+": "+"Enter name: ")
    unique_birthyears_dictionary[i]=final_name
````

````python
dirty_phonenumber_list = list()

with open('recweekdata.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            pass
        else:
            dirty_phonenumber_list.append(row[6])
        line_count += 1
        
# dirty_phonenumber_list

unique_phonenumbers = set(dirty_phonenumber_list)

# unique_phonenumbers

# [n for n in unique_phonenumbers]

# unique_phonenumbers_dictionary = {name:None for name in unique_phonenumbers}

# unique_phonenumbers_dictionary

phonenumbers_no_space = [n.replace(" ", "") for n in unique_phonenumbers] # removed spaces

# phonenumbers_no_space

# clean_phonenumber_list = list(a)

# clean_phonenumber_list

clean_phonenumber_list = []

for a in phonenumbers_no_space: # added +63 to the beginning of those without it
    if len(a) == 13:
        a = a
        clean_phonenumber_list.append(a)
    elif len(a) < 13:
        a = "+63" + a
        clean_phonenumber_list.append(a)
        
clean_phonenumber_list
````
